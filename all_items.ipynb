```python
import requests
import pandas
import time
```

```python
v2_url = 'https://api.guildwars2.com/v2'
items_url = f"{v2_url}/items"
```

```python
all_item_ids = [str(id) for id in requests.get(items_url).json()]
len(all_item_ids)
```

```python
ids = ",".join(all_item_ids[:200])
```

```python
df = pandas.DataFrame()
i = 0
while i < len(all_item_ids):
    ids = ",".join(all_item_ids[i:i+200])
    r = requests.get(items_url, params={"ids": ids})
    items = r.text
    df = df.append(pandas.read_json(items))
    i += 200
    if i % 5000 == 0:
        time.sleep(1)
```
